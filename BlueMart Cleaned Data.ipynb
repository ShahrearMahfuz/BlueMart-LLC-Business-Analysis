```python
df_Promo=pd.read_csv('bm_promotions.csv')
df_Promo.head()
```

```python
df_Promo.info()
df_Promo.duplicated().sum()
```

```python
df_Promo['start_date']=pd.to_datetime(df_Promo['start_date'])
df_Promo['end_date']=pd.to_datetime(df_Promo['end_date'])
```

```python
df_Promo.info()
df_Promo.duplicated().sum()
```

```python
##inserting promo_id column in sales table
for idx, promo in df_Promo.iterrows():
    mask = (
        (df_Sales['date'] >= promo['start_date']) &
        (df_Sales['date'] <= promo['end_date']) &
        (df_Sales['discount_pct'] == promo['discount_pct'])
    )
    df_Sales.loc[mask, 'promo_id'] = promo['promo_id']
```

```python
df_Sales.head()
```

```python
df_Sales.to_csv('sales_final.csv', index=False)
df_Promo.to_csv('promo_final.csv', index=False)
```
